```python
import pandas as pd
import numpy as np
import tensorflow
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('feature_table2.csv')
data.head()
```

```python
data['timestamp'] = data['timestamp'].astype(int)
data = data.set_index('timestamp')
cols = [c for c in data.columns if 'boottime' not in c]
data = data[cols]
data.head()
```

```python
data.tail()
```

```python
periods = []
broken = True
for i in data.index:
    if broken:
        start = i
        broken = False
    else:
        end = i
    if i+1 not in data.index:
        periods.append((start, end))
        broken = True
        print("continuous between ({}, {}) for {} timesteps".format(start, end, end-start+1))
```

```python
stride = 10
mark_points = [i for p in periods for i in range(p[0]+stride, p[1]+1)]
assert len(mark_points) == data.shape[0] - len(periods) * stride
```

```python
periods_a = [(34593, 34630), (11970, 12045), (28408, 28453), (80707, 80872)]
abnormal = [i for p in periods_a for i in range(p[0], p[1]+1) if i in mark_points]
normal = [i for i in mark_points if i not in abnormal]
```

```python
from random import shuffle
shuffle(normal)
shuffle(abnormal)
len_n = len(normal)
len_a = len(abnormal)
train_idx = normal[:int(0.6*len_n)]
val_n1_idx = normal[int(0.6*len_n):int(0.75*len_n)]
val_n2_idx = normal[int(0.75*len_n):int(0.9*len_n)]
test_n_idx = normal[int(0.9*len_n):]
val_a_idx = abnormal[:int(0.6*len_a)]
test_a_idx = abnormal[int(0.6*len_a):]
```

```python
train = data.loc[train_idx, :]
val_n1 = data.loc[val_n1_idx, :]
val_n2 = data.loc[val_n2_idx, :]
test_n = data.loc[test_n_idx, :]
val_a = data.loc[val_a_idx, :]
test_a = data.loc[test_a_idx, :]
```
